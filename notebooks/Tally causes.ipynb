```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df = pd.read_csv("../data/crash-merged-lat-lng.csv")
```

```python
f = open("../data/appendix-b.txt")
context = None
dd = dict()
for line in f:
    line = line.rstrip()
    if line[0] == 'x':
        context = " ".join(line.split()[1::])
        dd[ line.split()[0].replace("x","") ] = context
        print "Context is now %s" % context
    else:
        line = line.split()
        code = line[0]
        desc = " ".join(line[1::])
        print code, desc
        dd[code] = context + ":  " + desc
f.close()
```

```python
g = open("../data/causes.csv", "wb")
g.write("cause\n")
unknown = set()
for row in df.iterrows():
    vector = str(row[1]["CAUSES"]).lower().replace("a",""). \
        replace("b","").replace("c","").replace("d","").split()
    for elem in vector:
        if elem in dd:
            g.write('"' + dd[elem].replace('"',"'") + '"' + "\n")
            #g.write('"' + elem + '"' + "\n")
        else:
            unknown.add(elem)
g.close()
```

```python
g = open("../data/dates.csv", "wb")
g.write("date\n")
for elem in df["CRASH DATE"]:
    elem = elem.split("/")
    g.write( elem[2] + "-" + elem[1] + "\n")
g.close()
```

```python
g = open("../data/objects.csv", "wb")
g.write("object\n")
ignores=set()
for elem in df["OBJECTS STRUCK"]:
    if str(elem).lower() != "nan":
        for letter in elem:
            if letter in oo:
                g.write( '"' + oo[letter] + '"' + "\n")
            else:
                ignores.add(letter)
g.close()
            
```

```python
ww = dict({
    'F':'Fine',
    'M':'Mist/fog',
    'L':'Light rain',
    'H':'Heavy rain',
    'S':'Snow',
    'F':'Frost',
    'S':'Strong wind'
})
```

```python
g = open("../data/weathers.csv", "wb")
g.write("weather\n")
c = 0
for elem in df["WTHRa"]:
    elem = elem.replace(" ","")
    if len(elem) == 0:
        continue
    st = []
    for letter in elem:
        st.append(ww[letter])
        if elem == 'F':
            c += 1
    #print "and".join(st)
    g.write( " and ".join(st).lower() + "\n" )
g.close()
```
