```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df = pd.read_csv("../data/crash-merged-lat-lng.csv")
```

```python
dd = dict()
f = open("../data/appendix-b.csv")
for line in f:
    line = line.rstrip().split(",")
    dd[ line[0] ] = line[1]
f.close()
```

```python
dd
```

```python
g = open("/tmp/causes.txt", "wb")
for row in df.iterrows():
    g.write( str(row[1]["CAUSES"]) + "\n" )
g.close()
```

```python
counter = Counter()
for row in df.iterrows():
    vector = str(row[1]["CAUSES"]).lower()
    vector = vector.replace("a","").replace("b","").replace("c","").replace("d","")
    vector = vector.split()
    for elem in vector:
        if elem in dd:
            counter[ dd[elem] ] += 1
```

```python
g = open("../data/crash-tally.csv", "wb")
g.write("cause,count\n")
for key in counter:
    g.write( '"' + key + '",' + str(counter[key]) + "\n" )
g.close()
```
